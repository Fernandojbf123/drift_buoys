```python
datos_de_topografia = cargar_datos_de_topografia()
datos_de_topografia = recortar_datos_de_topografia_a_area_de_estudio(datos_de_topografia, -98, -90, 18, 26)
datos_de_topografia = reemplazar_valores_de_topografia(datos_de_topografia, valor_a_reemplazar=-10, nuevo_valor=-10)
datos_de_batimetria = cargar_datos_de_batimetria()
```

```python
carpeta = "C:\\Users\\Atmosfera\\Downloads\\20260219_doris"
archivos = [
    "datos_Localizacion_4876178_TOTAL.csv",
    "datos_Localizacion_4878152_TOTAL.csv",
    "datos_Localizacion_4878214_TOTAL.csv",
    "datos_Localizacion_4878225_TOTAL.csv",
    "datos_Localizacion_4878319_TOTAL.csv",
    "datos_Localizacion_4878504_TOTAL.csv",
    "datos_Localizacion_4878213_TOTAL.csv",
    "datos_Localizacion_4878218_TOTAL.csv",
    "datos_Localizacion_4878221_TOTAL.csv",
    "datos_Localizacion_4876190_TOTAL.csv"
]

```

```python
datos = {}
for archivo in archivos:
    os.path.join(carpeta, archivo)
    df = pd.read_csv(os.path.join(carpeta, archivo))
    serial = archivo.split("_")[2]
    print(f"Serial: {serial}")
    datos[serial] = df
```

```python
df = datos["4876178"]
fecha_ref_ini = pd.to_datetime("2026-02-20 21:00", format="%Y-%m-%d %H:%M")
fecha_ref_fin = pd.to_datetime("2026-02-20 22:00", format="%Y-%m-%d %H:%M")
df["tspan"] = pd.to_datetime(df["fecha"], format="%Y-%m-%dT%H:%M:%S.%fz")
df_filtrado = df[(df["tspan"] >= fecha_ref_ini) & (df["tspan"] <= fecha_ref_fin)]

lat = df_filtrado["latitud"]
lon = df_filtrado["logitud"]
tspan = df_filtrado["tspan"]

fig, ax = plt.subplots(figsize=(16, 7), subplot_kw={'projection': ccrs.PlateCarree()})
graficar_mapa_cartopy(ax, -93, -91, 18, 20)
graficar_mapa_topografico(axe=ax, topografia=datos_de_topografia, lon_min=-93, lon_max=-91, lat_min=18, lat_max=20)
graficar_batimetria_en_mapa(ax, datos_de_batimetria = datos_de_batimetria)
ax.scatter(lon, lat, c = "red", s=10)
```
